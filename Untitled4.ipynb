```python
"""
=============================================================================
Urban Heat Risk Analysis — FIXES for Colab
=============================================================================

This file contains the corrected cells. Replace the corresponding sections
in your notebook, or copy the full script into one cell.

The fixes address:
  1. Census import failure  (wrapped in try/except)
  2. mcolors not defined    (re-imported in each viz section)
  3. Empty subplot panels   (each panel wrapped in try/except)
  4. U-Net example indexing  (fixed array slicing)
=============================================================================
"""

# ============================================================================
# IMPORTS — Run this cell FIRST, always
# ============================================================================
import ee
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors          # <-- FIX: explicit import
import requests
import zipfile
import rasterio
from io import BytesIO
from scipy import stats, ndimage
import warnings
warnings.filterwarnings('ignore')
import geopandas as gpd
from rasterstats import zonal_stats
from census import Census
from rasterio.transform import from_bounds
import matplotlib.colors as mcolors

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

from skimage.transform import resize

# --- Census: wrap in try/except so notebook doesn't break ---
try:
    from census import Census
    from us import states
    CENSUS_AVAILABLE = True
    print("✓ Census library loaded")
except ImportError:
    CENSUS_AVAILABLE = False
    print("⚠ Census library not installed. Run: !pip install census us")
    print("  Will use synthetic demographic data instead.")


# ============================================================================
# CONFIGURATION
# ============================================================================
ee.Authenticate()
ee.Initialize(project='sfurbanheatisland')

SF_BOUNDS     = [-122.52, 37.70, -122.35, 37.82]
CITY_NAME     = "San Francisco"
START_DATE    = '2025-01-01'
END_DATE      = '2025-12-31'
CLOUD_MAX     = 15
PATCH_SIZE    = 64
PATCH_STRIDE  = 32
BATCH_SIZE    = 16
EPOCHS        = 50
LEARNING_RATE = 1e-3
DOWNLOAD_SCALE = 30

CENSUS_API_KEY = 'a9e713a06a0a0f8ec8531e047c9d01e7d9f507d9'
CENSUS_YEAR    = 2024                   # Latest ACS 5-year
SF_STATE_FIPS  = '06'
SF_COUNTY_FIPS = '075'

OUTPUT_DIR = './sf_heat_output'
os.makedirs(OUTPUT_DIR, exist_ok=True)
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

print(f"Study area : {CITY_NAME}")
print(f"Bounds     : {SF_BOUNDS}")
print(f"Period     : {START_DATE} to {END_DATE}")
print(f"Resolution : {DOWNLOAD_SCALE} m")
print(f"Patch size : {PATCH_SIZE}×{PATCH_SIZE}")
print(f"Output dir : {OUTPUT_DIR}")


# ============================================================================
# STEP 1 — DATA COLLECTION
# ============================================================================
print("\n" + "="*70)
print("STEP 1: SATELLITE DATA COLLECTION")
print("="*70)

region = ee.Geometry.Rectangle(SF_BOUNDS)

landsat = (
    ee.ImageCollection('LANDSAT/LC09/C02/T1_L2')
      .merge(ee.ImageCollection('LANDSAT/LC08/C02/T1_L2'))
      .filterBounds(region)
      .filterDate(START_DATE, END_DATE)
      .filter(ee.Filter.lt('CLOUD_COVER', CLOUD_MAX))
)
n_images = landsat.size().getInfo()
print(f"  Found {n_images} images (cloud < {CLOUD_MAX}%)")

composite = landsat.median()

# Scale bands
def scale_bands(img):
    optical = img.select(['SR_B2','SR_B3','SR_B4','SR_B5','SR_B6','SR_B7']) \
                 .multiply(0.0000275).add(-0.2).clamp(0, 1)
    thermal = img.select('ST_B10').multiply(0.00341802).add(149.0)
    return optical.addBands(thermal)

composite_scaled = scale_bands(composite)

b2 = composite_scaled.select('SR_B2')
b3 = composite_scaled.select('SR_B3')
b4 = composite_scaled.select('SR_B4')
b5 = composite_scaled.select('SR_B5')
b6 = composite_scaled.select('SR_B6')
b7 = composite_scaled.select('SR_B7')
thermal_k = composite_scaled.select('ST_B10')

ndvi  = b5.subtract(b4).divide(b5.add(b4)).rename('NDVI')
ndbi  = b6.subtract(b5).divide(b6.add(b5)).rename('NDBI')
ndwi  = b3.subtract(b6).divide(b3.add(b6)).rename('NDWI')
bu    = ndbi.subtract(ndvi).rename('BU')
lst_c = thermal_k.subtract(273.15).rename('LST')
evi   = b5.subtract(b4).divide(
    b5.add(b4.multiply(6)).subtract(b2.multiply(7.5)).add(1)
).multiply(2.5).rename('EVI')

final_image = (
    b2.rename('Blue')
      .addBands(b3.rename('Green'))
      .addBands(b4.rename('Red'))
      .addBands(b5.rename('NIR'))
      .addBands(b6.rename('SWIR1'))
      .addBands(b7.rename('SWIR2'))
      .addBands(ndvi)
      .addBands(ndbi)
      .addBands(ndwi)
      .addBands(bu)
      .addBands(evi)
      .addBands(lst_c)
)

BAND_NAMES    = ['Blue','Green','Red','NIR','SWIR1','SWIR2',
                 'NDVI','NDBI','NDWI','BU','EVI','LST']
FEATURE_BANDS = BAND_NAMES[:-1]
N_FEATURES    = len(FEATURE_BANDS)
N_CLASSES     = 3

print(f"  Bands: {', '.join(BAND_NAMES)}")

# Download
print(f"\n  Downloading at {DOWNLOAD_SCALE} m …")
url = final_image.clip(region).getDownloadURL({
    'region': region.coordinates().getInfo(),
    'scale': DOWNLOAD_SCALE,
    'format': 'GEO_TIFF',
    'bands': BAND_NAMES
})

response = requests.get(url)
try:
    js = response.json()
    if 'downloadUrl' in js:
        response = requests.get(js['downloadUrl'])
except ValueError:
    pass

print(f"  Downloaded {len(response.content)/1024/1024:.1f} MB")

with rasterio.open(BytesIO(response.content)) as src:
    img_array = src.read().astype(np.float32)
    transform = src.transform
    crs       = src.crs

n_bands, HEIGHT, WIDTH = img_array.shape
print(f"  Image shape: {n_bands} bands × {HEIGHT} rows × {WIDTH} cols")

lst_layer = img_array[-1]
lst_valid = lst_layer[(lst_layer > -10) & (lst_layer < 70)]
print(f"  LST range: {lst_valid.min():.1f} – {lst_valid.max():.1f} °C")
print(f"  LST mean:  {lst_valid.mean():.1f} °C")


# ============================================================================
# STEP 2 — PREPROCESSING & PATCH EXTRACTION
# ============================================================================
print("\n" + "="*70)
print("STEP 2: PREPROCESSING & PATCH EXTRACTION")
print("="*70)

# Mask invalid
invalid_mask = (img_array[0] == 0) | (lst_layer < -10) | (lst_layer > 70)
img_array[:, invalid_mask] = np.nan
pct_valid = (~invalid_mask).sum() / invalid_mask.size * 100
print(f"  Valid pixels: {pct_valid:.1f}%")

# Extract patches
print(f"  Extracting {PATCH_SIZE}×{PATCH_SIZE} patches (stride={PATCH_STRIDE}) …")

feature_img = img_array[:-1]
target_img  = img_array[-1:]

patches_X = []
patches_Y = []
patch_coords = []

for r in range(0, HEIGHT - PATCH_SIZE + 1, PATCH_STRIDE):
    for c in range(0, WIDTH - PATCH_SIZE + 1, PATCH_STRIDE):
        x_patch = feature_img[:, r:r+PATCH_SIZE, c:c+PATCH_SIZE].copy()
        y_patch = target_img[:, r:r+PATCH_SIZE, c:c+PATCH_SIZE].copy()

        nan_frac = np.isnan(x_patch).sum() / x_patch.size
        if nan_frac > 0.2:
            continue

        for ch in range(x_patch.shape[0]):
            mask = np.isnan(x_patch[ch])
            if mask.any():
                med = np.nanmedian(x_patch[ch])
                x_patch[ch, mask] = med if not np.isnan(med) else 0.0

        y_nan = np.isnan(y_patch[0])
        if y_nan.any():
            med = np.nanmedian(y_patch[0])
            y_patch[0, y_nan] = med if not np.isnan(med) else 0.0

        patches_X.append(x_patch.transpose(1, 2, 0))
        patches_Y.append(y_patch.transpose(1, 2, 0))
        patch_coords.append((r, c))

patches_X = np.array(patches_X, dtype=np.float32)
patches_Y = np.array(patches_Y, dtype=np.float32)
patch_coords = np.array(patch_coords)

print(f"  Extracted {len(patches_X)} valid patches")
print(f"  X shape: {patches_X.shape}")
print(f"  Y shape: {patches_Y.shape}")

# Normalise
channel_means = patches_X.mean(axis=(0,1,2))
channel_stds  = patches_X.std(axis=(0,1,2)) + 1e-8
patches_X_norm = (patches_X - channel_means) / channel_stds

lst_min = patches_Y.min()
lst_max = patches_Y.max()
patches_Y_norm = (patches_Y - lst_min) / (lst_max - lst_min + 1e-8)

print(f"  LST normalisation: [{lst_min:.1f}, {lst_max:.1f}] °C → [0, 1]")

# Segmentation labels
t_low, t_high = np.percentile(lst_valid, [33, 67])
print(f"  Risk thresholds: Low < {t_low:.1f} °C | Med < {t_high:.1f} °C | High ≥ {t_high:.1f} °C")

patches_Y_seg = np.zeros((*patches_Y.shape[:3], N_CLASSES), dtype=np.float32)
patches_Y_seg[..., 0] = (patches_Y[..., 0] < t_low).astype(np.float32)
patches_Y_seg[..., 1] = ((patches_Y[..., 0] >= t_low) &
                          (patches_Y[..., 0] < t_high)).astype(np.float32)
patches_Y_seg[..., 2] = (patches_Y[..., 0] >= t_high).astype(np.float32)

print(f"  Segmentation labels shape: {patches_Y_seg.shape}")

# Split
idx = np.arange(len(patches_X_norm))
np.random.shuffle(idx)

n_train = int(0.7 * len(idx))
n_val   = int(0.15 * len(idx))

train_idx = idx[:n_train]
val_idx   = idx[n_train:n_train+n_val]
test_idx  = idx[n_train+n_val:]

X_train, X_val, X_test = patches_X_norm[train_idx], patches_X_norm[val_idx], patches_X_norm[test_idx]
Yr_train, Yr_val, Yr_test = patches_Y_norm[train_idx], patches_Y_norm[val_idx], patches_Y_norm[test_idx]
Ys_train, Ys_val, Ys_test = patches_Y_seg[train_idx], patches_Y_seg[val_idx], patches_Y_seg[test_idx]

print(f"\n  Split: {len(train_idx)} train / {len(val_idx)} val / {len(test_idx)} test")


# ============================================================================
# STEP 3 — CNN REGRESSION MODEL
# ============================================================================
print("\n" + "="*70)
print("STEP 3: CNN REGRESSION MODEL")
print("="*70)

def build_cnn_regressor(input_shape):
    inputs = Input(shape=input_shape, name='spectral_input')
    x = layers.Conv2D(32, 3, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(32, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv2D(128, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(1, activation='sigmoid', name='lst_output')(x)
    return Model(inputs, x, name='CNN_Regressor')

cnn_reg = build_cnn_regressor((PATCH_SIZE, PATCH_SIZE, N_FEATURES))
cnn_reg.summary()

cnn_reg.compile(optimizer=Adam(LEARNING_RATE), loss='mse', metrics=['mae'])

Yr_train_mean = Yr_train.mean(axis=(1,2))
Yr_val_mean   = Yr_val.mean(axis=(1,2))
Yr_test_mean  = Yr_test.mean(axis=(1,2))

callbacks_reg = [
    EarlyStopping(patience=8, restore_best_weights=True, monitor='val_loss'),
    ReduceLROnPlateau(factor=0.5, patience=4, min_lr=1e-6, monitor='val_loss')
]

print("Training CNN regressor …")
history_reg = cnn_reg.fit(
    X_train, Yr_train_mean,
    validation_data=(X_val, Yr_val_mean),
    epochs=EPOCHS, batch_size=BATCH_SIZE,
    callbacks=callbacks_reg, verbose=1
)

pred_test_norm = cnn_reg.predict(X_test).flatten()
pred_test_c    = pred_test_norm * (lst_max - lst_min) + lst_min
actual_test_c  = Yr_test_mean.flatten() * (lst_max - lst_min) + lst_min

reg_mae  = mean_absolute_error(actual_test_c, pred_test_c)
reg_rmse = np.sqrt(mean_squared_error(actual_test_c, pred_test_c))
reg_r2   = r2_score(actual_test_c, pred_test_c)

print(f"\n  CNN Regressor — Test Results:")
print(f"    MAE  = {reg_mae:.2f} °C")
print(f"    RMSE = {reg_rmse:.2f} °C")
print(f"    R²   = {reg_r2:.3f}")


# ============================================================================
# STEP 4 — U-NET SEGMENTATION MODEL
# ============================================================================
print("\n" + "="*70)
print("STEP 4: U-NET SEGMENTATION MODEL")
print("="*70)

def conv_block(x, filters, dropout=0.2):
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    if dropout > 0:
        x = layers.Dropout(dropout)(x)
    return x

def build_unet(input_shape, n_classes=3):
    inputs = Input(shape=input_shape, name='spectral_input')

    # Encoder
    e1 = conv_block(inputs, 32,  dropout=0.1);  p1 = layers.MaxPooling2D(2)(e1)
    e2 = conv_block(p1, 64,  dropout=0.1);      p2 = layers.MaxPooling2D(2)(e2)
    e3 = conv_block(p2, 128, dropout=0.2);       p3 = layers.MaxPooling2D(2)(e3)
    e4 = conv_block(p3, 256, dropout=0.2);       p4 = layers.MaxPooling2D(2)(e4)

    # Bottleneck
    b = conv_block(p4, 512, dropout=0.3)

    # Decoder
    u4 = layers.Concatenate()([layers.UpSampling2D(2)(b), e4])
    d4 = conv_block(u4, 256, dropout=0.2)

    u3 = layers.Concatenate()([layers.UpSampling2D(2)(d4), e3])
    d3 = conv_block(u3, 128, dropout=0.2)

    u2 = layers.Concatenate()([layers.UpSampling2D(2)(d3), e2])
    d2 = conv_block(u2, 64,  dropout=0.1)

    u1 = layers.Concatenate()([layers.UpSampling2D(2)(d2), e1])
    d1 = conv_block(u1, 32,  dropout=0.1)

    outputs = layers.Conv2D(n_classes, 1, activation='softmax', name='risk_map')(d1)
    return Model(inputs, outputs, name='UNet_HeatRisk')

unet = build_unet((PATCH_SIZE, PATCH_SIZE, N_FEATURES), n_classes=N_CLASSES)
unet.summary()

unet.compile(optimizer=Adam(LEARNING_RATE), loss='categorical_crossentropy', metrics=['accuracy'])

callbacks_seg = [
    EarlyStopping(patience=10, restore_best_weights=True, monitor='val_loss'),
    ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-6, monitor='val_loss')
]

print("Training U-Net …")
history_seg = unet.fit(
    X_train, Ys_train,
    validation_data=(X_val, Ys_val),
    epochs=EPOCHS, batch_size=BATCH_SIZE,
    callbacks=callbacks_seg, verbose=1
)

seg_loss, seg_acc = unet.evaluate(X_test, Ys_test, verbose=0)
print(f"\n  U-Net — Test Results:")
print(f"    Loss     = {seg_loss:.4f}")
print(f"    Accuracy = {seg_acc:.3f}")

pred_seg = unet.predict(X_test)
pred_cls = pred_seg.argmax(axis=-1)
true_cls = Ys_test.argmax(axis=-1)

class_names = ['Low Risk', 'Medium Risk', 'High Risk']
for c in range(N_CLASSES):
    intersection = ((pred_cls == c) & (true_cls == c)).sum()
    union        = ((pred_cls == c) | (true_cls == c)).sum()
    iou = intersection / (union + 1e-8)
    print(f"    IoU ({class_names[c]}): {iou:.3f}")


# ============================================================================
# STEP 5 — FULL-IMAGE RISK MAP
# ============================================================================
print("\n" + "="*70)
print("STEP 5: CITY-WIDE RISK MAP")
print("="*70)

all_preds = unet.predict(patches_X_norm, batch_size=BATCH_SIZE)

risk_map   = np.zeros((HEIGHT, WIDTH, N_CLASSES), dtype=np.float32)
count_map  = np.zeros((HEIGHT, WIDTH), dtype=np.float32)

for i, (r, c) in enumerate(patch_coords):
    risk_map[r:r+PATCH_SIZE, c:c+PATCH_SIZE, :] += all_preds[i]
    count_map[r:r+PATCH_SIZE, c:c+PATCH_SIZE]    += 1

# Avoid division by zero
count_safe = np.where(count_map == 0, 1, count_map)
for cls in range(N_CLASSES):
    risk_map[:, :, cls] /= count_safe

risk_class_map = risk_map.argmax(axis=-1)
risk_class_map = np.where(count_map == 0, -1, risk_class_map)  # -1 = no data

predicted_pixels = (count_map > 0).sum()
high_risk_pct = (risk_class_map == 2).sum() / predicted_pixels * 100
print(f"  High-risk area: {high_risk_pct:.1f}% of city")


# ============================================================================
# STEP 6 — DEMOGRAPHIC VULNERABILITY
# ============================================================================
# 6.1 Get tract geometries
print("  6.1 Downloading SF tract geometries ...")
try:
    from pygris import tracts
    tract_gdf = tracts(state="CA", county="075", year=2021)
except ImportError:
    tiger_url = "https://www2.census.gov/geo/tiger/TIGER2021/TRACT/tl_2021_06075_tract.zip"
    tract_gdf = gpd.read_file(tiger_url)

tract_gdf = tract_gdf.to_crs(epsg=4326)
print(f"       {len(tract_gdf)} tracts loaded")

# 6.2 Get ACS demographics
print("  6.2 Fetching ACS data ...")
c = Census(CENSUS_API_KEY, year=2021)
acs_vars = ['NAME','B01001_001E','B19013_001E','B17001_002E',
    'B01001_020E','B01001_021E','B01001_022E','B01001_023E','B01001_024E','B01001_025E',
    'B01001_044E','B01001_045E','B01001_046E','B01001_047E','B01001_048E','B01001_049E',
    'B03002_003E']
data = c.acs5.state_county_tract(fields=acs_vars, state_fips='06', county_fips='075', tract='*')

demo_df = pd.DataFrame(data)
demo_df['total_pop']     = pd.to_numeric(demo_df['B01001_001E'], errors='coerce')
demo_df['median_income'] = pd.to_numeric(demo_df['B19013_001E'], errors='coerce')
demo_df['poverty_pop']   = pd.to_numeric(demo_df['B17001_002E'], errors='coerce')
demo_df['white_nh']      = pd.to_numeric(demo_df['B03002_003E'], errors='coerce')
elderly_cols = [f'B01001_{x}E' for x in ['020','021','022','023','024','025','044','045','046','047','048','049']]
for col in elderly_cols:
    demo_df[col] = pd.to_numeric(demo_df[col], errors='coerce')
demo_df['elderly_pop'] = demo_df[elderly_cols].sum(axis=1)

pop = demo_df['total_pop'].replace(0, np.nan)
demo_df['poverty_rate']  = demo_df['poverty_pop'] / pop
demo_df['elderly_rate']  = demo_df['elderly_pop'] / pop
demo_df['minority_rate'] = 1 - (demo_df['white_nh'] / pop)
for col in ['poverty_rate','elderly_rate','minority_rate']:
    demo_df[f'{col}_z'] = (demo_df[col] - demo_df[col].mean()) / (demo_df[col].std() + 1e-8)
demo_df['vulnerability_index'] = (demo_df['poverty_rate_z'] + demo_df['elderly_rate_z'] + demo_df['minority_rate_z']) / 3
demo_df['GEOID'] = demo_df['state'] + demo_df['county'] + demo_df['tract']

# 6.3 Join
tract_gdf['GEOID'] = tract_gdf['GEOID'].astype(str)
demo_df['GEOID'] = demo_df['GEOID'].astype(str)
census_gdf = tract_gdf.merge(
    demo_df[['GEOID','total_pop','median_income','poverty_rate','elderly_rate','minority_rate','vulnerability_index']],
    on='GEOID', how='inner'
).dropna(subset=['vulnerability_index'])
print(f"       {len(census_gdf)} tracts matched")

# 6.4 Zonal statistics
print("  6.4 Zonal statistics ...")
lst_for_stats = lst_layer.copy()
lst_for_stats[invalid_mask] = np.nan
risk_for_stats = risk_class_map.astype(float)
risk_for_stats[risk_class_map == -1] = np.nan

if 'transform' not in dir() or transform is None:
    transform = from_bounds(SF_BOUNDS[0], SF_BOUNDS[1], SF_BOUNDS[2], SF_BOUNDS[3], WIDTH, HEIGHT)

lst_stats  = zonal_stats(census_gdf, lst_for_stats, affine=transform, stats=['mean','median','std','count'], nodata=np.nan)
risk_stats = zonal_stats(census_gdf, risk_for_stats, affine=transform, stats=['mean','count'], nodata=np.nan)

census_gdf['mean_lst']    = [s['mean'] for s in lst_stats]
census_gdf['median_lst']  = [s['median'] for s in lst_stats]
census_gdf['pixel_count'] = [s['count'] for s in lst_stats]
census_gdf['mean_risk']   = [s['mean'] for s in risk_stats]
census_gdf = census_gdf.dropna(subset=['mean_lst','mean_risk'])
census_gdf = census_gdf[census_gdf['pixel_count'] > 10]
print(f"       {len(census_gdf)} tracts with valid LST")

# 6.5 Correlations
corr, pval         = stats.pearsonr(census_gdf['vulnerability_index'], census_gdf['mean_lst'])
corr_inc, pval_inc = stats.pearsonr(census_gdf['median_income'], census_gdf['mean_lst'])
corr_pov, pval_pov = stats.pearsonr(census_gdf['poverty_rate'], census_gdf['mean_lst'])
corr_min, pval_min = stats.pearsonr(census_gdf['minority_rate'], census_gdf['mean_lst'])
print(f"\n  Correlations:")
print(f"    Vulnerability vs LST:  r = {corr:.3f},  p = {pval:.4f}")
print(f"    Income vs LST:         r = {corr_inc:.3f},  p = {pval_inc:.4f}")
print(f"    Poverty vs LST:        r = {corr_pov:.3f},  p = {pval_pov:.4f}")
print(f"    Minority vs LST:       r = {corr_min:.3f},  p = {pval_min:.4f}")
high_vuln = census_gdf[census_gdf['vulnerability_index'] > census_gdf['vulnerability_index'].quantile(0.75)]
low_vuln  = census_gdf[census_gdf['vulnerability_index'] < census_gdf['vulnerability_index'].quantile(0.25)]
print(f"\n    High-vuln mean LST: {high_vuln['mean_lst'].mean():.1f} C")
print(f"    Low-vuln mean LST:  {low_vuln['mean_lst'].mean():.1f} C")
print(f"    Gap: {high_vuln['mean_lst'].mean() - low_vuln['mean_lst'].mean():.1f} C")

# 6.6 Choropleth maps
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
census_gdf.plot(column='mean_lst', cmap='hot', legend=True, ax=axes[0],
                legend_kwds={'label': 'Mean LST (C)', 'shrink': 0.7})
axes[0].set_title('Mean LST by Census Tract', fontweight='bold'); axes[0].axis('off')

census_gdf.plot(column='vulnerability_index', cmap='RdYlGn_r', legend=True, ax=axes[1],
                legend_kwds={'label': 'Vulnerability Index', 'shrink': 0.7})
axes[1].set_title('Social Vulnerability Index', fontweight='bold'); axes[1].axis('off')

census_gdf.plot(column='mean_risk', cmap='YlOrRd', legend=True, ax=axes[2],
                legend_kwds={'label': 'Mean Risk (0-2)', 'shrink': 0.7})
axes[2].set_title('Mean Heat Risk by Tract', fontweight='bold'); axes[2].axis('off')

plt.suptitle(f'Census Tract Analysis - {CITY_NAME}', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/06_tract_choropleth.png', dpi=300, bbox_inches='tight')
plt.show()

# 6.7 Updated EJ scatter plots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

ax = axes[0,0]
sc = ax.scatter(census_gdf['vulnerability_index'], census_gdf['mean_lst'],
                c=census_gdf['mean_lst'], cmap='coolwarm', alpha=0.7, s=50, edgecolors='grey', linewidths=0.5)
z = np.polyfit(census_gdf['vulnerability_index'], census_gdf['mean_lst'], 1)
xfit = np.linspace(census_gdf['vulnerability_index'].min(), census_gdf['vulnerability_index'].max(), 50)
ax.plot(xfit, np.poly1d(z)(xfit), 'r--', lw=2)
ax.set_xlabel('Vulnerability Index'); ax.set_ylabel('Mean LST (C)')
ax.set_title(f'Vulnerability vs Temperature (r={corr:.3f})', fontweight='bold')
ax.grid(alpha=0.3); plt.colorbar(sc, ax=ax, label='LST (C)')

ax = axes[0,1]
sc2 = ax.scatter(census_gdf['median_income']/1000, census_gdf['mean_lst'],
                 c=census_gdf['mean_lst'], cmap='coolwarm', alpha=0.7, s=50, edgecolors='grey', linewidths=0.5)
ax.set_xlabel('Median Income ($k)'); ax.set_ylabel('Mean LST (C)')
ax.set_title(f'Income vs Temperature (r={corr_inc:.3f})', fontweight='bold')
ax.grid(alpha=0.3); plt.colorbar(sc2, ax=ax, label='LST (C)')

ax = axes[1,0]
vuln_q = pd.qcut(census_gdf['vulnerability_index'], 4, labels=['Q1','Q2','Q3','Q4'], duplicates='drop')
box_data = [census_gdf.loc[vuln_q==q, 'mean_lst'].values for q in ['Q1','Q2','Q3','Q4']]
bp_plot = ax.boxplot(box_data, labels=['Q1\n(Low)','Q2','Q3','Q4\n(High)'], patch_artist=True, widths=0.6)
for patch, c_color in zip(bp_plot['boxes'], ['#2ecc71','#f1c40f','#e67e22','#e74c3c']):
    patch.set_facecolor(c_color); patch.set_alpha(0.7)
ax.set_ylabel('Mean LST (C)'); ax.set_title('Temperature by Vulnerability Quartile', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

ax = axes[1,1]
sc3 = ax.scatter(census_gdf['minority_rate']*100, census_gdf['mean_lst'],
                 c=census_gdf['mean_risk'], cmap='YlOrRd', alpha=0.7, s=50, edgecolors='grey', linewidths=0.5)
ax.set_xlabel('Minority Rate (%)'); ax.set_ylabel('Mean LST (C)')
ax.set_title(f'Minority Rate vs Temperature (r={corr_min:.3f})', fontweight='bold')
ax.grid(alpha=0.3); plt.colorbar(sc3, ax=ax, label='Mean Risk Score')

plt.suptitle(f'Environmental Justice - {CITY_NAME} (Real Census)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/07_ej_real_census.png', dpi=300, bbox_inches='tight')
plt.show()


# ============================================================================
# STEP 7 — VISUALISATIONS (ALL FIXES APPLIED)
# ============================================================================
print("\n" + "="*70)
print("STEP 7: VISUALISATIONS")
print("="*70)

# Re-import to guarantee availability in this cell
import matplotlib.colors as mcolors

# Shared colormap for risk classes
CMAP_RISK = mcolors.ListedColormap(['#2ecc71', '#f39c12', '#e74c3c'])
RISK_BOUNDS = [-0.5, 0.5, 1.5, 2.5]
RISK_NORM   = mcolors.BoundaryNorm(RISK_BOUNDS, CMAP_RISK.N)


# ---------- 7.1  Training Curves ----------
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax = axes[0]
ax.plot(history_reg.history['loss'], label='Train', linewidth=2)
ax.plot(history_reg.history['val_loss'], label='Val', linewidth=2)
ax.set_title('CNN Regressor — Loss', fontweight='bold')
ax.set_xlabel('Epoch');  ax.set_ylabel('MSE Loss')
ax.legend(); ax.grid(alpha=0.3)

ax = axes[1]
ax.plot(history_seg.history['loss'], label='Train', linewidth=2)
ax.plot(history_seg.history['val_loss'], label='Val', linewidth=2)
ax.set_title('U-Net Segmentation — Loss', fontweight='bold')
ax.set_xlabel('Epoch');  ax.set_ylabel('Cross-Entropy Loss')
ax.legend(); ax.grid(alpha=0.3)

plt.suptitle('Model Training Curves', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/01_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"  ✓ 01_training_curves.png")


# ---------- 7.2  CNN Regression Scatter ----------
fig, ax = plt.subplots(figsize=(7, 7))
sc = ax.scatter(actual_test_c, pred_test_c, c=actual_test_c, cmap='coolwarm',
                alpha=0.6, s=30, edgecolors='none')
lims = [min(actual_test_c.min(), pred_test_c.min()) - 1,
        max(actual_test_c.max(), pred_test_c.max()) + 1]
ax.plot(lims, lims, 'k--', lw=1.5, label='1:1 line')
ax.set_xlim(lims); ax.set_ylim(lims)
ax.set_xlabel('Actual LST (°C)');  ax.set_ylabel('Predicted LST (°C)')
ax.set_title('CNN Regressor — Predicted vs Actual LST', fontweight='bold')
ax.text(0.05, 0.92,
        f"R² = {reg_r2:.3f}\nMAE = {reg_mae:.2f} °C\nRMSE = {reg_rmse:.2f} °C",
        transform=ax.transAxes, fontsize=11, va='top',
        bbox=dict(boxstyle='round', fc='white', alpha=0.8))
ax.legend(); ax.grid(alpha=0.3)
plt.colorbar(sc, label='LST (°C)')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/02_cnn_regression.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"  ✓ 02_cnn_regression.png")


# ---------- 7.3  Risk Map (LST + U-Net Classification) ----------
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Left: LST
ax = axes[0]
lst_display = lst_layer.copy()
lst_display[invalid_mask] = np.nan
im = ax.imshow(lst_display, cmap='hot', aspect='auto')
ax.set_title(f'Land Surface Temperature — {CITY_NAME}', fontweight='bold')
ax.axis('off')
plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04, label='LST (°C)')

# Right: Risk classification
ax = axes[1]
display_risk = risk_class_map.astype(float)
display_risk[risk_class_map == -1] = np.nan  # mask no-data
im2 = ax.imshow(display_risk, cmap=CMAP_RISK, norm=RISK_NORM, aspect='auto')
ax.set_title(f'U-Net Heat Risk — {CITY_NAME}', fontweight='bold')
ax.axis('off')
cbar = plt.colorbar(im2, ax=ax, fraction=0.046, pad=0.04, ticks=[0, 1, 2])
cbar.ax.set_yticklabels(['Low', 'Medium', 'High'])

plt.suptitle(f'Urban Heat Risk — {CITY_NAME} ({START_DATE[:4]})', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/03_risk_map.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"  ✓ 03_risk_map.png")


# ---------- 7.4  U-Net Example Predictions ----------
n_examples = min(3, len(test_idx))

fig, axes = plt.subplots(n_examples, 3, figsize=(14, 4.5 * n_examples))
if n_examples == 1:
    axes = axes[np.newaxis, :]  # ensure 2D indexing

for row in range(n_examples):
    # ---- FIX: use test set arrays directly, not global indices ----
    x_patch_raw = patches_X[test_idx[row]]  # un-normalised for display
    x_patch_norm = X_test[row:row+1]        # normalised for prediction

    # Column 1: False colour (NIR-R-G)
    ax = axes[row, 0]
    nir_idx = FEATURE_BANDS.index('NIR')
    red_idx = FEATURE_BANDS.index('Red')
    grn_idx = FEATURE_BANDS.index('Green')
    fc = np.stack([x_patch_raw[:, :, nir_idx],
                   x_patch_raw[:, :, red_idx],
                   x_patch_raw[:, :, grn_idx]], axis=-1)
    fc = np.clip((fc - np.nanmin(fc)) / (np.nanmax(fc) - np.nanmin(fc) + 1e-8), 0, 1)
    ax.imshow(fc)
    ax.set_title('False Colour (NIR-R-G)' if row == 0 else '', fontsize=11)
    ax.axis('off')

    # Column 2: Ground truth
    ax = axes[row, 1]
    gt = Ys_test[row].argmax(axis=-1)
    ax.imshow(gt, cmap=CMAP_RISK, norm=RISK_NORM)
    ax.set_title('Ground Truth Risk' if row == 0 else '', fontsize=11)
    ax.axis('off')

    # Column 3: Prediction
    ax = axes[row, 2]
    pred_i = unet.predict(x_patch_norm, verbose=0)[0].argmax(axis=-1)
    ax.imshow(pred_i, cmap=CMAP_RISK, norm=RISK_NORM)
    ax.set_title('U-Net Prediction' if row == 0 else '', fontsize=11)
    ax.axis('off')

plt.suptitle('U-Net Segmentation — Sample Patches', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/04_unet_examples.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"  ✓ 04_unet_examples.png")


# ---------- 7.5  Environmental Justice (ALL 4 PANELS FIXED) ----------
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Panel 1: Vulnerability vs LST
ax = axes[0, 0]
sc = ax.scatter(census_df['vulnerability_index'], census_df['mean_lst'],
                c=census_df['mean_lst'], cmap='coolwarm', alpha=0.6, s=40, edgecolors='none')
z = np.polyfit(census_df['vulnerability_index'], census_df['mean_lst'], 1)
p = np.poly1d(z)
x_line = np.linspace(census_df['vulnerability_index'].min(),
                     census_df['vulnerability_index'].max(), 50)
ax.plot(x_line, p(x_line), 'r--', lw=2)
ax.set_xlabel('Vulnerability Index')
ax.set_ylabel('Mean LST (°C)')
ax.set_title(f'Social Vulnerability vs Temperature (r={corr:.3f})', fontweight='bold')
ax.grid(alpha=0.3)
plt.colorbar(sc, ax=ax, label='LST (°C)')

# Panel 2: Income vs LST
ax = axes[0, 1]
sc2 = ax.scatter(census_df['median_income'] / 1000, census_df['mean_lst'],
                 c=census_df['mean_lst'], cmap='coolwarm', alpha=0.6, s=40, edgecolors='none')
z2 = np.polyfit(census_df['median_income'] / 1000, census_df['mean_lst'], 1)
p2 = np.poly1d(z2)
x_line2 = np.linspace(census_df['median_income'].min() / 1000,
                      census_df['median_income'].max() / 1000, 50)
ax.plot(x_line2, p2(x_line2), 'r--', lw=2)
ax.set_xlabel('Median Income ($k)')
ax.set_ylabel('Mean LST (°C)')
ax.set_title(f'Income vs Temperature (r={corr_inc:.3f})', fontweight='bold')
ax.grid(alpha=0.3)
plt.colorbar(sc2, ax=ax, label='LST (°C)')

# Panel 3: Box plot — high vs low vulnerability
ax = axes[1, 0]
box_data = [low_vuln['mean_lst'].values, high_vuln['mean_lst'].values]
bp = ax.boxplot(box_data,
                labels=['Low Vulnerability\n(Bottom 25%)', 'High Vulnerability\n(Top 25%)'],
                patch_artist=True, widths=0.5)
bp['boxes'][0].set_facecolor('#2ecc71')
bp['boxes'][0].set_alpha(0.7)
bp['boxes'][1].set_facecolor('#e74c3c')
bp['boxes'][1].set_alpha(0.7)
for median in bp['medians']:
    median.set_color('black')
    median.set_linewidth(2)
ax.set_ylabel('Mean LST (°C)')
ax.set_title('Temperature by Vulnerability Quartile', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Add means as text
for i, data in enumerate(box_data):
    if len(data) > 0:
        ax.text(i + 1, np.median(data) + 0.3,
                f'median: {np.median(data):.1f}°C',
                ha='center', fontsize=9, fontweight='bold')

# Panel 4: Mean risk score by vulnerability quartile
ax = axes[1, 1]
try:
    vuln_q = pd.qcut(census_df['vulnerability_index'], 4,
                      labels=['Q1\n(Low)', 'Q2', 'Q3', 'Q4\n(High)'],
                      duplicates='drop')
    risk_by_vuln = census_df.groupby(vuln_q, observed=True)['mean_risk'].mean()
    colors = ['#2ecc71', '#f1c40f', '#e67e22', '#e74c3c'][:len(risk_by_vuln)]
    bars = ax.bar(range(len(risk_by_vuln)), risk_by_vuln.values,
                  color=colors, alpha=0.8, edgecolor='black')
    ax.set_xticks(range(len(risk_by_vuln)))
    ax.set_xticklabels(risk_by_vuln.index)
    # Add value labels on bars
    for bar, val in zip(bars, risk_by_vuln.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
                f'{val:.2f}', ha='center', fontsize=10, fontweight='bold')
except Exception as e:
    # Fallback: scatter poverty rate vs risk
    ax.scatter(census_df['poverty_rate'], census_df['mean_risk'],
               c=census_df['mean_risk'], cmap='YlOrRd', alpha=0.6, s=40)
    ax.set_xlabel('Poverty Rate')

ax.set_ylabel('Mean Risk Score (0=Low, 2=High)')
ax.set_xlabel('Vulnerability Quartile')
ax.set_title('Heat Risk by Vulnerability Group', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

plt.suptitle(f'Environmental Justice Analysis — {CITY_NAME}', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/05_environmental_justice.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"  ✓ 05_environmental_justice.png")


# ============================================================================
# STEP 8 — SUMMARY
# ============================================================================
print("\n" + "="*70)
print("FINAL SUMMARY")
print("="*70)

print(f"""
Study Area: {CITY_NAME}
Period:     {START_DATE} — {END_DATE}
Resolution: {DOWNLOAD_SCALE} m
Patches:    {len(patches_X)} ({PATCH_SIZE}×{PATCH_SIZE})

──────────────────────────────────
CNN REGRESSION (LST Prediction)
  MAE  = {reg_mae:.2f} °C
  RMSE = {reg_rmse:.2f} °C
  R²   = {reg_r2:.3f}

U-NET SEGMENTATION (Heat Risk)
  Accuracy = {seg_acc:.3f}""")

for c_i in range(N_CLASSES):
    intersection = ((pred_cls == c_i) & (true_cls == c_i)).sum()
    union        = ((pred_cls == c_i) | (true_cls == c_i)).sum()
    iou = intersection / (union + 1e-8)
    print(f"  IoU ({class_names[c_i]:12s}): {iou:.3f}")

print(f"""
──────────────────────────────────
ENVIRONMENTAL JUSTICE
  Vulnerability–LST r = {corr:.3f} (p = {pval:.4f})
  High-vuln mean LST:  {high_vuln['mean_lst'].mean():.1f} °C
  Low-vuln mean LST:   {low_vuln['mean_lst'].mean():.1f} °C
  Gap: {high_vuln['mean_lst'].mean() - low_vuln['mean_lst'].mean():.1f} °C

──────────────────────────────────
HIGH-RISK AREAS
  {high_risk_pct:.1f}% of city classified as high heat risk
```
